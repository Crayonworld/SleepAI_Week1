```python
import wfdb
import matplotlib.pyplot as plt

# 
record = wfdb.rdrecord('datasets/apnea-ecg/a01')

print(record.__dict__.keys())  # 
print(record.sig_name)         # 
print(record.fs)               # 

#  10  ECG 
plt.plot(record.p_signal[:record.fs*10,0])
plt.title("ECG Signal - First 10 seconds")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```
