```python
import wfdb
import matplotlib.pyplot as plt

# 读取一条记录
record = wfdb.rdrecord('datasets/apnea-ecg/a01')

print(record.__dict__.keys())  # 查看包含哪些数据
print(record.sig_name)         # 信号名称
print(record.fs)               # 采样率

# 绘制前 10 秒的 ECG 信号
plt.plot(record.p_signal[:record.fs*10,0])
plt.title("ECG Signal - First 10 seconds")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```
